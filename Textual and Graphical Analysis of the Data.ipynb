```python
#  Function to calculate chances of being promoted or not based on certain criteria
def chances_of_promotion(criterion = None):
    
    value_list = []
    
    promoted_total_length = data[data['Promoted_or_Not'] == 1].count().max()
    
    for value in data[criterion]:
        
        if value in value_list:
            
            pass
        
        else:
            
            value_list.append(value)
    
    
    for criterion_value in value_list: 
        
        print('--------------------------')
        
        print(criterion_value)
        
        print('Percentage - {} '.format(((data[(data[criterion] == criterion_value) & (data['Promoted_or_Not'] == 1)].count().max()/promoted_total_length)*100)))
    
```
